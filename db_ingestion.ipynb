```python
# 1. parsing dataset
filepath = 'Dataset.xls'
orders_df = pd.read_excel(filepath, sheet_name='Orders')
returns_df = pd.read_excel(filepath, sheet_name='Returns')
people_df = pd.read_excel(filepath, sheet_name='People')

# 2. creating local database
conn = sqlite3.connect('Superstore.db')

# 3. transfroming python tables into sql tables in db
orders_df.to_sql('Orders', conn, if_exists='replace', index=False)
returns_df.to_sql('Returns', conn, if_exists='replace', index=False)
people_df.to_sql('People', conn, if_exists='replace', index=False)

print("SQL Database is ready to run!")
```

```python
query = """
SELECT 
    Category,
    ROUND(SUM(Sales), 2) AS Total_Sales,
    ROUND(SUM(Profit), 2) AS Total_Profit,
    ROUND((SUM(Profit) / SUM(Sales)) * 100, 2) AS Profit_Margin_Percent
FROM Orders
GROUP BY Category
ORDER BY Total_Profit DESC;
"""

# Running query and printing results
result_df = pd.read_sql_query(query, conn)
result_df
```
